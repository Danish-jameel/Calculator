```python
import math

# Define the calculator functions
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error: Division by zero is not allowed."
    return x / y

def power(x, y):
    return x ** y

def sqrt(x):
    if x < 0:
        return "Error: Cannot calculate square root of a negative number."
    return math.sqrt(x)

def log(x, base=10):
    if x <= 0:
        return "Error: Logarithm is undefined for zero or negative numbers."
    return math.log(x, base)

def factorial(x):
    if x < 0:
        return "Error: Factorial of a negative number is undefined."
    return math.factorial(x)

def sin(x):
    return math.sin(math.radians(x))

def cos(x):
    return math.cos(math.radians(x))

def tan(x):
    try:
        return math.tan(math.radians(x))
    except:
        return "Error: Undefined for tan(90° + n*180°)."

def scientific_calculator():
    print("Scientific Calculator")
    print("---------------------")
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Power")
    print("6. Square Root")
    print("7. Logarithm (base 10)")
    print("8. Logarithm (custom base)")
    print("9. Factorial")
    print("10. Sine")
    print("11. Cosine")
    print("12. Tangent")

    while True:
        # Take input from the user
        choice = input("\nEnter choice (1-12): ")

        if choice in ['1', '2', '3', '4', '5']:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))

            if choice == '1':
                print(f"{num1} + {num2} = {add(num1, num2)}")
            elif choice == '2':
                print(f"{num1} - {num2} = {subtract(num1, num2)}")
            elif choice == '3':
                print(f"{num1} * {num2} = {multiply(num1, num2)}")
            elif choice == '4':
                print(f"{num1} / {num2} = {divide(num1, num2)}")
            elif choice == '5':
                print(f"{num1} ^ {num2} = {power(num1, num2)}")

        elif choice == '6':
            num = float(input("Enter number: "))
            print(f"√{num} = {sqrt(num)}")

        elif choice == '7':
            num = float(input("Enter number: "))
            print(f"log10({num}) = {log(num)}")

        elif choice == '8':
            num = float(input("Enter number: "))
            base = float(input("Enter base: "))
            print(f"log{base}({num}) = {log(num, base)}")

        elif choice == '9':
            num = int(input("Enter number: "))
            print(f"{num}! = {factorial(num)}")

        elif choice == '10':
            angle = float(input("Enter angle (in degrees): "))
            print(f"sin({angle}) = {sin(angle)}")

        elif choice == '11':
            angle = float(input("Enter angle (in degrees): "))
            print(f"cos({angle}) = {cos(angle)}")

        elif choice == '12':
            angle = float(input("Enter angle (in degrees): "))
            print(f"tan({angle}) = {tan(angle)}")

        else:
            print("Invalid Input")

        # Ask if the user wants another calculation
        next_calculation = input("\nDo you want to perform another calculation? (yes/no): ")
        if next_calculation.lower() != "yes":
            break

# Call the scientific calculator function
scientific_calculator()
```
